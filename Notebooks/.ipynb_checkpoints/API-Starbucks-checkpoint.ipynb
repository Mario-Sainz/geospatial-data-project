```python
donde = "madrid"
def geocode(address):
    data = requests.get(f"https://geocode.xyz/{address}?json=1").json()
    try:
        return {
            "type":"Point",
            "coordinates":[float(data["longt"]),float(data["latt"])]}
    except:
        return data
```

```python
from functools import reduce
import operator
```

```python
mapa_nombre = ["venue","name"]
latitud = ["venue", "location","lat"]
longitud = ["venue","location","lng"]
```

```python
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
list_Starbucks = []
for diccionario in Starbucks:
    paralista = {}
    paralista["name"] = getFromDict(diccionario,mapa_nombre)
    paralista["latitud"] = getFromDict(diccionario,latitud)
    paralista["longitud"] = getFromDict(diccionario,longitud)
    list_Starbucks.append(paralista)
```

```python
Starbucks1 = pd.DataFrame(list_Starbucks)
```

```python
Starbucks = Starbucks1.rename(columns = {'name': 'type'}, inplace = False)
```
