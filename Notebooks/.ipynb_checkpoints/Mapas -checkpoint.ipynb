```python
import requests
import json
from dotenv import load_dotenv
import os
import pandas as pd
from functools import reduce
import operator
from pymongo import MongoClient
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
```

```python
companies_madrid = pd.read_csv('companies_madrid.csv')
```

```python
Starbucks = pd.read_csv('Starbucks.csv')
```

```python
clubs = pd.read_csv('clubs.csv')
```

```python
condiciones = pd.concat([companies_madrid , Starbucks, clubs], axis=0, ignore_index=True)
```

```python
condiciones
```

```python
inicial_lat = 40.423863
inicial_long = -3.685858
```

```python
map_1 = folium.Map(location = [inicial_lat,inicial_long], zoom_start = 15)
```

```python
mi_empresa = Marker(location=[inicial_lat,inicial_long], tooltip = "Mi empresa")
mi_empresa.add_to(map_1)
```

```python
for i,row in condiciones.iterrows():
    ubicaciones = {
        "location" : [row["latitud"], row["longitud"]],
        "tooltip" : row["type"]
    }
    if row["type"] == "company":
        icon = Icon( color = "red",
                    prefix = "fa",
                    icon = "briefcase",
                    icon_color = "black"
        )
    elif row["type"] == "clubs":
        icon = Icon( color = "black",
                    prefix = "fa",
                    icon = "glass",
                    icon_color = "white"
        )
    else:
        icon = Icon( color = "green",
                    prefix = "fa",
                    icon = "coffee",
                    icon_color = "black"
        )
    Marker (**ubicaciones,icon = icon).add_to(map_1)
```
