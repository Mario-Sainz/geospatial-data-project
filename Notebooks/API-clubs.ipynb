```python
donde = "madrid"
def geocode(address):
    data = requests.get(f"https://geocode.xyz/{address}?json=1").json()
    try:
        return {
            "type":"Point",
            "coordinates":[float(data["longt"]),float(data["latt"])]}
    except:
        return data
```

```python
tok1 = os.getenv("tok1")
tok2 = os.getenv("tok2")
```

```python
enlace = 'https://api.foursquare.com/v2/venues/explore'
```

```python
parametros = { "client_id" : tok1,
          "client_secret" : tok2,
          "v" : "20180323",
          "ll": f"{madrid.get('coordinates')[1]},{madrid.get('coordinates')[0]}",
          "query":"clubs"
             }
```

```python
response = requests.get(url = enlace, params = parametros)
```

```python
data_clubs = json.loads(response.text)
```

```python
data_clubs1 = data_clubs.get("response")
```

```python
data_clubs2 = data_clubs1.get("groups")[0]
```

```python
clubs = data_clubs2.get("items")
```

```python
clubs[0]
```

```python
from functools import reduce
import operator
```

```python
mapa_nombre = ["venue","name"]
latitud = ["venue", "location","lat"]
longitud = ["venue","location","lng"]
```

```python
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
list_clubs = []
for diccionario in clubs:
    paralista = {}
    paralista["name"] = getFromDict(diccionario,mapa_nombre)
    paralista["latitud"] = getFromDict(diccionario,latitud)
    paralista["longitud"] = getFromDict(diccionario,longitud)
    list_clubs.append(paralista)
```

```python
clubs = pd.DataFrame(list_clubs)
```

```python
clubs["type"]= "clubs"
```
